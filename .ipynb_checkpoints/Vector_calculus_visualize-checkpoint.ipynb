```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Z = np.sqrt(X**2 + Y**2)
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

```python
plt.plot(fig_size=(16,8))
plt.subplot(121)
plt.imshow(Z)
plt.subplot(122)
plt.contour(Z)
```

```python
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X,Y= np.meshgrid(X, Y)

fig= plt.figure(figsize=(24,8))
ax=plt.subplot(131)
q = ax.quiver(X, Y, X, Y)
plt.xlim([-8,8])
plt.ylim([-8,8])
ax=plt.subplot(132)
q = ax.quiver(X, Y, -Y, X)
plt.xlim([-8,8])
plt.ylim([-8,8])
ax=plt.subplot(133)
q = ax.quiver(X, Y, X-Y, Y+X)
plt.xlim([-8,8])
plt.ylim([-8,8])
plt.show()
```
