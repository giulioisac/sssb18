```python
import numpy as np
import scipy as sp
import sympy as sym
sym.init_printing()
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
#help(sym.Function)
```

```python
t, S, a_0, n, pi, L, b_n = sym.symbols('t S a_0 n pi L b_n')
f = sym.Function('f')(t)
f
```

```python
sum1 = sym.Sum(n, (n, 1, sym.oo))
sum1
```

```python
sym.Eq(S * f, a_0 + sum1)
```
