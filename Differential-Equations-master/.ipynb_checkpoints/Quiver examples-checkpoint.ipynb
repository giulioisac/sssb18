```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,Y = np.meshgrid(np.arange(0,2*np.pi,.5), np.arange(0,2*np.pi,.5) )
U = X ** 2 + np.sin(Y)
V = 3 * Y

#1
plt.figure(figsize=(16,16))
Q = plt.quiver( U, V)
qk = plt.quiverkey(Q, 0.5, 0.92, 2, r'$2 \frac{m}{s}$', labelpos='W',
               fontproperties={'weight': 'bold'})
l,r,b,t = plt.axis()
dx, dy = r-l, t-b
plt.axis([l-0.05*dx, r+0.05*dx, b-0.05*dy, t+0.05*dy])

plt.title('Minimal arguments, no kwargs')
plt.show()
```

```python
#2
plt.figure(figsize=(16,16))
Q = plt.quiver( X, Y, U, V, units='width')
qk = plt.quiverkey(Q, 0.9, 0.95, 2, r'$2 \frac{m}{s}$',
               labelpos='E',
               coordinates='figure',
               fontproperties={'weight': 'bold'})
plt.axis([-1, 7, -1, 7])
plt.title('scales with plot width, not view')

plt.show()
```

```python
#3
plt.figure(figsize=(16,16))
Q = plt.quiver( X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],
            pivot='mid', color='r', units='inches' )
qk = plt.quiverkey(Q, 0.5, 0.03, 1, r'$1 \frac{m}{s}$', fontproperties={'weight': 'bold'})
plt.plot( X[::3, ::3], Y[::3, ::3], 'k.')
plt.axis([-1, 7, -1, 7])
plt.title("pivot='mid'; every third arrow; units='inches'")

plt.show()
```

```python
4#
plt.figure(figsize=(16,16))
M = np.sqrt(np.power(U, 2) + np.power(V, 2))
Q = plt.quiver( X, Y, U, V, M, units='x', pivot='tip', width=0.022, scale=1/0.15)
qk = plt.quiverkey(Q, 0.9, 1.05, 1, r'$1 \frac{m}{s}$',
                            labelpos='E',
                            fontproperties={'weight': 'bold'})
plt.plot(X, Y, 'k.')
plt.axis([-1, 7, -1, 7])
plt.title("scales with x view; pivot='tip'")

plt.show()
```

```python
#5
plt.figure(figsize=(16,16))
Q = plt.quiver( X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],
             color='r', units='x',
            linewidths=(2,), edgecolors=('k'), headaxislength=5 )
qk = plt.quiverkey(Q, 0.5, 0.03, 1, r'$1 \frac{m}{s}$', fontproperties={'weight': 'bold'})
plt.axis([-1, 7, -1, 7])
plt.title("triangular head; scale with x view; black edges")

plt.show()
```

```python
#6
plt.figure(figsize=(16,16))
M = np.zeros(U.shape, dtype='bool')
M[U.shape[0]/3:2*U.shape[0]/3,U.shape[1]/3:2*U.shape[1]/3] = True
U = np.ma.masked_array(U, mask=M)
V = np.ma.masked_array(V, mask=M)
Q = plt.quiver( U, V)
qk = plt.quiverkey(Q, 0.5, 0.92, 2, r'$2 \frac{m}{s}$', labelpos='W',
               fontproperties={'weight': 'bold'})
l,r,b,t = plt.axis()
dx, dy = r-l, t-b
plt.axis([l-0.05*dx, r+0.05*dx, b-0.05*dy, t+0.05*dy])
plt.title('Minimal arguments, no kwargs - masked values')


plt.show()
```
